```python
import pandas as  pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\S.Chandrasekaran5\OneDrive - Shell\Documents\Work\PAM\Research\BSP\GT5_exp\SVM\Brunei_Champion_GT5Model_Testing.csv')
```

```python
list_columns = [col for col in df.columns if 'anomaly_ratio' in col]
```

```python
xtickmark =[]
for count in range(len(list_columns)):
    temp = 'Model_'+ str(count+1)
    xtickmark.append(temp)
```

```python
df[list_columns].plot.line(legend=xtickmark)
plt.legend(xtickmark,loc='best')
plt.grid()
plt.xlabel('Weekly Metrics')
plt.ylabel('Anomaly Ratio [No Unit]')
```
